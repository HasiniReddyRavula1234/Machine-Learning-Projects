```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import kagglehub
path = kagglehub.dataset_download("olafkrastovski/handwritten-digits-0-9")
```

```python
import zipfile

zip_path = "handwritten-digits-0-9.zip"
extract_path = "/content/digits"
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
```

```python
import tensorflow as tf
from tensorflow import keras
from keras import layers, models
```

```python
train_df = keras.utils.image_dataset_from_directory(
    directory = '/content/digits',
    batch_size=32,
    image_size = (28, 28),
    color_mode='grayscale',
    validation_split = 0.2,
    subset="training",
    seed=123
)
validation_df = keras.utils.image_dataset_from_directory(
    directory = '/content/digits',
    batch_size=32,
    image_size = (28, 28),
    color_mode='grayscale',
    validation_split = 0.2,
    subset="validation",
    seed=123
)
```

```python
def process(image, label):
  image = tf.cast(image/255. , tf.float32)
  return image, label

  train_df = train_df.map(process)
  validation_df = validation_df.map(process)
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size = (3, 3), activation = 'relu', input_shape = (28, 28, 1)))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(64, kernel_size = (3, 3), activation = 'relu', input_shape = (28, 28, 1)))
model.add(MaxPooling2D(pool_size = (2, 2)))

model.add(Conv2D(128, kernel_size = (3, 3), activation = 'relu', input_shape = (28, 28, 1)))
model.add(MaxPooling2D(pool_size = (2, 2)))


model.add(Flatten())

model.add(Dense(128, activation = 'relu'))
keras.layers.Dropout(0.5),
model.add(Dense(10, activation = 'softmax'))
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
history = model.fit(train_df, epochs = 10, validation_data= validation_df)
```

```python
model.evaluate(validation_df)
```
